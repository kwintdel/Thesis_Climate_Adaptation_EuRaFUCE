```python
times = pd.read_csv("../AOA_data/CLUSTER3_time_multiple2.csv").values.flatten()
averages = pd.read_csv("../AOA_data/CLUSTER3_avrgmean_multiple2.csv").values.flatten()

times2 = pd.read_csv("../AOA_data/CLUSTER3_time_multiple3.csv").values.flatten()
averages2 = pd.read_csv("../AOA_data/CLUSTER3_avrgmean_multiple3.csv").values.flatten()
```

```python

# Create the combined dataframe with appropriate labels
df1 = pd.DataFrame({'value': averages, 'label': '0.01'})
df2 = pd.DataFrame({'value': averages2, 'label': '0.1'})
df = pd.concat([df1, df2], ignore_index=True)

plt.figure(figsize=(8, 6))

# Use a custom color palette
box_palette = {'0.01': 'skyblue', '0.1': 'seagreen'}
strip_palette = {'0.01': 'steelblue', '0.1': 'darkgreen'}

# Plot the boxplot and stripplot
sns.boxplot(x='label', y='value', hue = 'label', data=df, width=0.5, showfliers=False, palette=box_palette)
sns.stripplot(x='label', y='value', hue = 'label', data=df, jitter=True, alpha=0.7, edgecolor='black', palette=strip_palette)

# Labels and layout
plt.xlabel('Sample percentage [%]', fontsize=15)
plt.ylabel('Normalized average distance', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Define custom palettes
violin_palette = {'0.01': 'skyblue', '0.1': 'seagreen'}
strip_palette = {'0.01': 'steelblue', '0.1': 'darkgreen'}  # slightly darker

# Violin plot
sns.violinplot(
    x='label', y='value',hue = 'label', data=df,
    inner=None, width=0.3, palette=violin_palette
)

# Strip plot
sns.stripplot(
    x='label', y='value',hue = 'label', data=df,
    jitter=True, alpha=0.7, edgecolor='black',
    palette=strip_palette
)

# Labels and layout
plt.xlabel('Sample percentage [%]', fontsize=15)
plt.ylabel('Normalized average distance', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()
```
